```python
import numpy as np

# Parámetros para las distribuciones
lambda_poisson = 1  # Parámetro de la distribución Poisson (número promedio de siniestros)
mean_gamma = 2000000  # Media de la distribución Gamma (coste promedio por siniestro)
shape_gamma = 8000  # Parámetro de forma de la distribución Gamma
scale_gamma = 250  # Parámetro de escala de la distribución Gamma

# Número de simulaciones
simulaciones = 100000


# Simulación del número de siniestros usando la distribución Poisson
siniestros = np.random.poisson(lambda_poisson, simulaciones)

# Simulación del coste por siniestro usando la distribución Gamma
costo_siniestro = np.random.gamma(shape_gamma, scale_gamma, simulaciones)

# Cálculo del coste total para cada simulación
costoTotal = siniestros * costo_siniestro

# Cálculo del coste total promedio
costopromedio = np.mean(costoTotal)

print(f"El coste total promedio simulado para el próximo año es de {costopromedio:.2f}")

```

```python
import matplotlib.pyplot as plt
plt.hist(costoTotal, bins=50, edgecolor='k', alpha=0.7)
plt.axvline(costopromedio, color='r', linestyle='dashed', linewidth=1)
plt.title('Distribución del Coste Total Simulado para el Próximo Año')
plt.xlabel('Coste Total')
plt.ylabel('Frecuencia')
plt.show()
```

```python
def calcular_valor_futuro(vp, tasa, periodo):
    """
    Calcula el valor futuro de una inversión.

    :param pv: Valor presente (inversión inicial)
    :param tasa_interes: Tasa de interés por período (en decimal, por ejemplo, 0.05 para 5%)
    :param num_periodos: Número de períodos
    :return: Valor futuro de la inversión
    """
    valor_futuro = vp * (1 + tasa) ** periodo
    return valor_futuro


vp = 1000  # Valor presente
tasa = 0.05  # Tasa de interés del 5%
periodo = 10  # Número de períodos

fv = calcular_valor_futuro(vp, tasa, periodo)
print(f"El valor futuro de la inversión es: {fv}")
```

```python
def calcular_valor_presente(vf, tasa_interes, num_periodos):

    valor_presente = vf / (1 + tasa_interes) ** num_periodos
    return valor_presente


fv = 1628.89  # Valor futuro
tasa_interes = 0.05  # Tasa de interés del 5%
num_periodos = 10  # Número de períodos

pv = calcular_valor_presente(fv, tasa_interes, num_periodos)
print(f"El valor presente de la inversión es: {pv}")
```

```python
def convertir_tasa_nominal_a_efectiva(tasa_nominal, num_periodos):

    tasa_efectiva = (1 + tasa_nominal / num_periodos) ** num_periodos - 1
    return tasa_efectiva


tasa_nominal = 0.34  # Tasa nominal del 12%
num_periodos = 12  # Capitalización mensual

tasa_efectiva = convertir_tasa_nominal_a_efectiva(tasa_nominal, num_periodos)
print(f"La tasa efectiva anual es: {tasa_efectiva:.4f}")
```

```python
def convertir_tasa_efectiva_a_nominal(tasa_efectiva, num_periodos):
   
    tasa_nominal = num_periodos * ((1 + tasa_efectiva) ** (1 / num_periodos) - 1)
    return tasa_nominal


tasa_efectiva = 0.36  # Tasa efectiva del 36%
num_periodos = 4  # Capitalización mensual

tasa_nominal = convertir_tasa_efectiva_a_nominal(tasa_efectiva, num_periodos)
print(f"La tasa nominal anual es: {tasa_nominal:}")
```

```python
def convertir_tasa_efectiva_anual_a_nominal_mensual(tasa_efectiva_anual):
    
    num_periodos = 12  # Capitalización mensual
    tasa_nominal_mensual = num_periodos * ((1 + tasa_efectiva_anual) ** (1 / num_periodos) - 1)
    return tasa_nominal_mensual

# Ejemplo de uso
tasa_efectiva_anual = 0.4116  # Tasa efectiva del 12.68%

tasa_nominal_mensual = convertir_tasa_efectiva_anual_a_nominal_mensual(tasa_efectiva_anual)
print(f"La tasa nominal mensual es: {tasa_nominal_mensual:.4f}")
```
