# Ejercicio 1

La media de la cartera de seguros nos dice que el comportamiento de la siniestralidad para el producto de automóviles sigue una Distribución Poisson con Lambda igual a 1 y una severidad (coste en promedio) que sigue una Distribución Gamma con media igual a 2 millones y varianza igual a 5 millones. Simule el costo de la cartera para el próximo año. 

```python
# Librerías 
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Simulación de la distribución Poisson 

# Parámetro lambda
lambda_poisson = 1

# Simulación de 1000 valores aleatorios
poisson = np.random.poisson(lambda_poisson, 1000)
poisson
```

```python
# Distribución Poisson 
plt.hist(poisson)
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.title('Distribución Poisson')
plt.show()
```

```python
# Simulación de la distribución Gamma

# Media = k*theta = 2000000
# Varianza = k*theta^2 = 5000000
# Así, theta = 2.5 y k = 800000

# Parámetros 
k = 800000  
theta = 2.5

# Simulación de 1000 valores aleatorios
gamma = np.random.gamma(k, theta, 1000)
gamma
```

```python
# Distribución gamma
plt.hist(gamma)
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.title('Distribución Gamma')
plt.show()
```

```python
# Suma producto de un ejemplo 
coste = np.sum(poisson * gamma)
coste
```

```python
# Ahora realizaremos n ensayos del coste 

n = 1000 
n_distribucion = 1000 
lambda_poisson = 1
k = 800000  
theta = 2.5

coste_simulacion = []

for i in range(n):
    poisson = np.random.poisson(lambda_poisson, n_distribucion)
    gamma = np.random.gamma(k, theta, n_distribucion)
    coste = np.sum(poisson * gamma)
    coste_simulacion.append(coste)

# Promedio 
promedio = np.mean(coste_simulacion)
promedio 
```

```python
# Distribución coste
plt.hist(coste_simulacion)
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.title('Distribución Gamma')
plt.show()
```

```python
import numpy as np

n = 1000  
lambda_poisson = 1  
k = 800000  
theta = 2.5  


coste_simulacion = []

for i in range(n):
    num_siniestros = np.random.poisson(lambda_poisson)
    
    if num_siniestros > 0:
        # Generar la severidad de cada siniestro según la distribución Gamma
        gamma_costes = np.random.gamma(k, theta, num_siniestros)
        
        # Calcular el coste total para esta simulación del año
        coste_total = np.sum(gamma_costes)
    else:
        coste_total = 0
    
    # Guardar el coste total en la lista de simulación
    coste_simulacion.append(coste_total)

# Promedio del coste anual
promedio = np.mean(coste_simulacion)
promedio
```

# Ejercicio 3

Se desea financiar un inmueble que tiene un valor comercial de 300.000.000 de pesos. Se sabe que para un crédito hipotecario el valor máximo de préstamo es del 70% del valor total del inmueble y que la tasa de interés es del 13% efectivo anual. Calcule el valor de la couta de pago del crédito teniendo en cuenta el monto máximo de financiación y que este se financió con un plazo de 20 años. 
Con base en las funciones anteriormente desarrolladas en Python, realice la tabla de amortización en Python. 

```python
import pandas as pd
pd.options.display.float_format = '{:,.2f}'.format
pd.set_option('display.max_rows', None)
```

```python
def tasa_mensual(tasa_interes):
    tasa_mensual = (1 + tasa_interes) ** (1/12) - 1
    return tasa_mensual
```

```python
def tabla_amortizacion(valor, porcentaje, tasa_interes, plazo):
    # Monto del préstamo
    monto = valor * porcentaje

    # tasa de interés mensual
    tasa =tasa_mensual(tasa_interes)

    # número total de pagos 
    n = plazo * 12

    # cuota mensual 
    cuota_mensual = monto * tasa / (1 - (1 + tasa) ** (-n))

    # Tabla de amortización
    saldo_pendiente = monto
    tabla_amortizacion = []

    # Generar la tabla de amortización
    for mes in range(1, n + 1):
        interes_mes = saldo_pendiente * tasa
        amortizacion_mes = cuota_mensual - interes_mes
        saldo_pendiente -= amortizacion_mes
        tabla_amortizacion.append([mes, interes_mes, amortizacion_mes, cuota_mensual, saldo_pendiente])

    df_amortizacion = pd.DataFrame(tabla_amortizacion, columns=["Mes", "Intereses", "Amortización", "Cuota Total", "Saldo Pendiente"])
    return df_amortizacion
```

```python
valor = 300000000  
porcentaje = 0.70  
tasa_interes = 0.13  
plazo = 20  

tabla_amortizacion(valor, porcentaje, tasa_interes, plazo)
```
