```python
#Ejercicio 1
# Función para calcular la probabilidad de supervivencia
def probabilidad_supervivencia(qx):
    return 1 - qx

# Función para calcular la probabilidad de fallecimiento
def probabilidad_fallecimiento(qx):
    return qx

# Probabilidad de fallecimiento antes de los 51 años
q50 = 0.015

# Llamadas a las funciones
prob_survivir_hasta_51 = probabilidad_supervivencia(q50)
prob_fallecer_antes_51 = probabilidad_fallecimiento(q50)

# Mostrar los resultados
print("Probabilidad de que una persona de 50 años sobreviva hasta los 51 años:", prob_survivir_hasta_51)
print("Probabilidad de que una persona de 50 años fallezca antes de los 51 años:", prob_fallecer_antes_51)
```
