# Ejercicio 1

En una tabla de mortalidad simplificada, se da la siguiente información para una persona de 50 años. Hay una probabilidad de 1.5% de que la persona de 50 años fallezca antes de cumplir 51. Preguntas: Cuál es la probabilidad de que una persona de 50 años sobreviva hasta los 51 años

```python
def calcular_probabilidades(p65, p66=None):
    """
    Calcula la probabilidad de que una persona de 65 años fallezca antes de los 66 y
    la probabilidad de que sobreviva hasta los 67 años.
    
    :param p65: Probabilidad de que una persona de 65 años sobreviva hasta los 66.
    :param p66: Probabilidad de que una persona de 66 años sobreviva hasta los 67. Si no se proporciona, se asume igual a p65.
    :return: Un diccionario con las probabilidades solicitadas.
    """
    # Si no se proporciona p66, se asume que es igual a p65
    if p66 is None:
        p66 = p65
    
    # Probabilidad de fallecer antes de los 66 años
    prob_fallecer_antes_66 = 1 - p65
    
    # Probabilidad de sobrevivir hasta los 67 años
    prob_sobrevivir_hasta_67 = p65 * p66
    
    return {
        "fallecer_antes_66": prob_fallecer_antes_66,
        "sobrevivir_hasta_67": prob_sobrevivir_hasta_67
    }

# Ejemplo de uso con p65 = 0.96
resultado = calcular_probabilidades(0.96)
print("Probabilidad de fallecer antes de los 66 años:", resultado["fallecer_antes_66"])
print("Probabilidad de sobrevivir hasta los 67 años:", resultado["sobrevivir_hasta_67"])
```

```python
def calcular_probabilidades_supervivencia(p70, p71, p72):
    """
    Calcula la probabilidad de que una persona de 70 años sobreviva hasta los 72 años
    y la probabilidad de que fallezca antes de cumplir los 72 años.

    :param p70: Probabilidad de que una persona de 70 años sobreviva hasta los 71.
    :param p71: Probabilidad de que una persona de 71 años sobreviva hasta los 72.
    :param p72: Probabilidad de que una persona de 72 años sobreviva hasta los 73.
    :return: Un diccionario con las probabilidades solicitadas.
    """
    # Probabilidad de sobrevivir hasta los 72 años
    prob_sobrevivir_hasta_72 = p70 * p71
    
    # Probabilidad de fallecer antes de cumplir los 72 años
    prob_fallecer_antes_72 = 1 - prob_sobrevivir_hasta_72
    
    return {
        "sobrevivir_hasta_72": prob_sobrevivir_hasta_72,
        "fallecer_antes_72": prob_fallecer_antes_72
    }

# Ejemplo de uso con los valores de la tabla
resultado = calcular_probabilidades_supervivencia(0.95, 0.92, 0.90)
print("Probabilidad de sobrevivir hasta los 72 años:", resultado["sobrevivir_hasta_72"])
print("Probabilidad de fallecer antes de los 72 años:", resultado["fallecer_antes_72"])
```

```python
def calcular_probabilidades_mortalidad(p60, q61):
    """
    Calcula la probabilidad de fallecimiento antes de los 61 años (q60) y la probabilidad
    de que una persona de 60 años sobreviva hasta los 62 años.

    :param p60: Probabilidad de que una persona de 60 años sobreviva hasta los 61.
    :param q61: Probabilidad de fallecimiento entre los 61 y 62 años.
    :return: Un diccionario con las probabilidades solicitadas.
    """
    # Probabilidad de fallecer antes de los 61 años (q60)
    q60 = 1 - p60
    
    # Probabilidad de sobrevivir hasta los 62 años
    p61 = 1 - q61
    prob_sobrevivir_hasta_62 = p60 * p61
    
    return {
        "fallecer_antes_61": q60,
        "sobrevivir_hasta_62": prob_sobrevivir_hasta_62
    }

# Ejemplo de uso con los valores dados en el ejercicio
resultado = calcular_probabilidades_mortalidad(0.97, 0.035)
print("Probabilidad de fallecer antes de los 61 años (q60):", resultado["fallecer_antes_61"])
print("Probabilidad de sobrevivir hasta los 62 años:", resultado["sobrevivir_hasta_62"])
```

- Calcule la renta vitalicia luego de la jubilación a partir del ahorro. 

```python
import pandas as pd

# Función para generar la tabla de ahorro mes a mes
def generar_tabla_ahorro_mensual(salario, tasa_aporte, tasa_interes_mensual, meses):
    aporte_mensual = salario * tasa_aporte  # Aporte ya mensual
    saldo_acumulado = 0
    
    tabla_ahorro = []

    for mes in range(1, meses + 1):
        saldo_acumulado = (saldo_acumulado + aporte_mensual) * (1 + tasa_interes_mensual)
        tabla_ahorro.append({"Mes": mes, "Aporte Mensual": int(aporte_mensual), "Saldo Acumulado": int(saldo_acumulado)})

    return pd.DataFrame(tabla_ahorro)

# Parámetros
salario = 7000000
tasa_aporte = 0.16
tasa_interes_mensual = 0.04 / 12  # Tasa mensual ya calculada
meses_aporte = 32 * 12  # 32 años en meses
AA=pd.DataFrame(tabla_ahorro_mensual)

reserva_actuarial=salario*tasa_aporte*meses_aporte

# Generar la tabla de ahorro mensual
tabla_ahorro_mensual = generar_tabla_ahorro_mensual(salario, tasa_aporte, tasa_interes_mensual, meses_aporte)
#Renta_vitalicia= tabla_ahorro_mensual[1]*meses_aporte

# Mostrar la tabla
print(tabla_ahorro_mensual)
print("Reserva Actuarial;",reserva_actuarial)
```
