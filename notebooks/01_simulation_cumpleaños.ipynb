```python
import random
import matplotlib.pyplot as plt

def has_duplicate_birthdays(num_people):
    """Simula los cumpleaños de num_people personas y verifica si hay duplicados."""
    birthdays = [random.randint(1, 365) for _ in range(num_people)]
    return len(birthdays) != len(set(birthdays))

def monte_carlo_birthday_paradox(num_people, num_simulations):
    """Simula la paradoja del cumpleaños usando Monte Carlo y devuelve la probabilidad de cumpleaños compartidos."""
    duplicate_count = 0
    
    for _ in range(num_simulations):
        if has_duplicate_birthdays(num_people):
            duplicate_count += 1
    
    # Calcula la probabilidad de que al menos dos personas compartan el mismo cumpleaños
    probability = duplicate_count / num_simulations
    return probability, duplicate_count, num_simulations - duplicate_count

# Parámetros
num_people = 52  # Número de personas en la habitación
num_simulations = 10000  # Número de simulaciones a realizar

# Ejecuta la simulación de Montecarlo
probability, duplicate_count, non_duplicate_count = monte_carlo_birthday_paradox(num_people, num_simulations)

# Muestra el resultado
print(f"En un grupo de {num_people} personas, la probabilidad de que dos compartan cumpleaños es aproximadamente {probability:.2%}")

# Datos para el gráfico de barras
casos = ['Success', 'Failure']
cantidad = [duplicate_count, non_duplicate_count]

# Crear el gráfico de barras
plt.figure(figsize=(10, 6))
plt.bar(casos, cantidad, color=['green', 'red'])

# Añadir títulos y etiquetas
plt.title('% of success /Failure')
plt.xlabel('Case')
plt.ylabel('Amount')

# Añadir etiquetas encima de las barras con porcentajes
for i, v in enumerate(cantidad):
    percentage = (v / num_simulations) * 100
    plt.text(i, v + 0.5, f'{percentage:.2f}%', ha='center', va='bottom')

# Mostrar el gráfico
plt.show()
```

```python
import random
import matplotlib.pyplot as plt

def has_duplicate_birthdays(num_people):
    """Simula los cumpleaños de num_people personas y verifica si hay duplicados."""
    birthdays = [random.randint(1, 365) for _ in range(num_people)]
    return len(birthdays) != len(set(birthdays))

def monte_carlo_birthday_paradox(num_people, num_simulations):
    """Simula la paradoja del cumpleaños usando Monte Carlo y devuelve la probabilidad de cumpleaños compartidos."""
    duplicate_count = 0
    
    for _ in range(num_simulations):
        if has_duplicate_birthdays(num_people):
            duplicate_count += 1
    
    # Calcula la probabilidad de que al menos dos personas compartan el mismo cumpleaños
    probability = duplicate_count / num_simulations
    return probability

# Parámetros
num_simulations = 10000  # Número de simulaciones a realizar
max_people = 70  # Número máximo de personas en la habitación

# Almacenar probabilidades para diferentes números de personas
probabilities = []

# Ejecuta la simulación de Montecarlo para diferentes números de personas
for num_people in range(1, max_people + 1):
    probability = monte_carlo_birthday_paradox(num_people, num_simulations)
    probabilities.append(probability)

# Crear el gráfico de la curva
plt.figure(figsize=(10, 6))
plt.plot(range(1, max_people + 1), probabilities, marker='o', linestyle='-', color='b')

# Añadir títulos y etiquetas
plt.title('Probability')
plt.xlabel('Number of people in the experiment')
plt.ylabel('Probability')

# Mostrar el gráfico
plt.grid(True)
plt.show()
```
